```python
data=pd.read_excel('C:/Users/HP/Desktop/Data Science/project/0520.xlsx')
data['股票代码']=data['股票代码'].astype(object)
data.head(10)
```

```python
data['市盈率']=data['市盈率'].fillna(-1)
data['现价']=data['现价'].fillna(data['现价'].mean())
data['涨跌幅']=data['涨跌幅'].fillna(data['涨跌幅'].mean())
data['涨跌']=data['涨跌'].fillna(data['涨跌'].mean())
data['涨速']=data['涨速'].fillna(data['涨速'].mean())
data['换手']=data['换手'].fillna(data['换手'].mean())
data['量比']=data['量比'].fillna(data['量比'].mean())
data['振幅']=data['振幅'].fillna(data['振幅'].mean())
```

```python
def str2value(valueStr):
    valueStr = str(valueStr)
    idxOfYi = valueStr.find('亿')
    idxOfWan = valueStr.find('万')
    if idxOfYi != -1 and idxOfWan != -1:
        return int(float(valueStr[:idxOfYi])*1e8 + float(valueStr[idxOfYi+1:idxOfWan])*1e4)
    elif idxOfYi != -1 and idxOfWan == -1:
        return int(float(valueStr[:idxOfYi])*1e8)
    elif idxOfYi == -1 and idxOfWan != -1:
        return int(float(valueStr[idxOfYi+1:idxOfWan])*1e4)
    elif idxOfYi == -1 and idxOfWan == -1:
        return float(valueStr)
```

```python
trade=[]
for x in data['成交额']:
    trade.append(str2value(x))

data.insert(loc=9,column='成交额x',value=trade)
```

```python
current=[]
for x in data['流通股']:
    current.append(str2value(x))

data.insert(loc=11,column='流通股x',value=current)
```

```python
marvalue=[]
for x in data['流通市值']:
    marvalue.append(str2value(x))

data.insert(loc=13,column='流通市值x',value=marvalue)
```

```python
ratio=[]
for x in data['市盈率']:
    if x=='亏损':
        ratio.append(0)
    else:
        ratio.append(float(x))

data.insert(loc=15,column='市盈率x',value=ratio)
```

```python
data['成交额x']=data['成交额x'].fillna(data['成交额x'].mean())
data['流通股x']=data['流通股x'].fillna(data['流通股x'].mean())
data['流通市值x']=data['流通市值x'].fillna(data['流通市值x'].mean())
```
